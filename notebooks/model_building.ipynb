Build a regression model.

```python
import pandas as pd
import statsmodels.api as sm
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Reset display options to default values
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
df_chosen_var = pd.read_csv('../data/df_chosen_var.csv')
df_chosen_var
```

```python
# Removing 1 row with a NaN value 
df_chosen_var_cleaned = df_chosen_var.dropna()
df_chosen_var_cleaned
```

```python
# Create a function that will run a regression model with the x variables as inputs
def regression_modeling(x_vars):
    X = df_chosen_var_cleaned[x_vars]
    y = df_chosen_var_cleaned['number_of_bikes_avail']

    # Add a constant to the model
    X = sm.add_constant(X)

    # Create the model
    model = sm.OLS(y, X)

    # Fit the model
    results = model.fit()

    return results.summary()
```

```python
# First model with all the x variables from Foursquare dataset
x_variables = ['average_distance_fs', 'average_price_fs', 'average_popularity_fs', 'average_rating_fs']
print(regression_modeling(x_variables))
```

```python
# Second model with all the x variables from Yelp dataset
x_variables = ['average_distance_yelp', 'average_price_yelp', 'average_popularity_yelp', 'average_rating_yelp']
print(regression_modeling(x_variables))
```
